### 2. Feature Engineering

From 2 feature 'variation_value' và 'variation_type', the 'variation_type' have 6 type of Size, Color, Scent, Type, Size + Concentration + Formulation, Size + Concentration and 'variation_value' presenting the accordiing size, color or scent... 5 feature are engineered as:
* oz: Extract the oz size of the product
* color: The color of the product
* scent: The scent of the product
* type: The type of the product
* quantity_included: the quantity included in the product, Ex: 1 pack of face wipes have 10 wipes

The color, scent, type features are the label encoded

The null values are then filled with -1 to indicate that the product does not have this variation

```python
def extract_oz(value):
    match = re.search(r'(\d*\.?\d+)\s*oz', value.strip())
    return float(match.group(1)) if match else -1

df['oz'] = df['variation_value'].apply(extract_oz)
```

```python
df['scent_raw'] = df.apply(
    lambda row: row['variation_value'] if row['variation_type'] == 'Scent' else None,
    axis=1
)

df['scent'] = None
non_null_scent = df['scent_raw'].notnull()
df.loc[non_null_scent, 'scent'] = encoder.fit_transform(df.loc[non_null_scent, 'scent_raw'])
df.drop(columns=['scent_raw'], inplace=True)
```

```python
df['color_raw'] = df.apply(
    lambda row: row['variation_value'] if row['variation_type'] == 'Color' else None,
    axis=1
)
valid_colors = {
    'Rose', 'Blush', 'Original', 'Peppermint', 'Vanilla Beige', 'Cool Dewy', 'Vanilla',
    'Berry', 'Minted Rose Lip Balm', 'Rosebud Salve', 'Rosebud Salve in a Tube',
    'Burgundy', 'Transforming Neutral Beige', 'Translucent', 'Fuchsia', 'Jade', 'Clear',
    'Coconut', 'Green', 'Pink', 'Watermelon', 'Translucent Tone 1', 'Pearl Pink',
    'Cucumber', 'Light', 'Light/Medium', 'Fair - Light', 'White', 'Petal Pink', 'Caramel',
    'Honey', 'Stone', 'White/Off-white', 'Navy', 'Fete', '772 Classic', 'Bliss',
    'Peach Of Cake!', 'Lavender', 'New York'
}
df['color_raw'] = df['color_raw'].where(df['color_raw'].isin(valid_colors))


df['color'] = None
non_null_color = df['color_raw'].notnull()
df.loc[non_null_color, 'color'] = encoder.fit_transform(df.loc[non_null_color, 'color_raw'])
df.drop(columns=['color_raw'], inplace=True)
```

```python
df['type_raw'] = df.apply(
    lambda row: row['variation_value'] if row['variation_type'] == 'Type' else None,
    axis=1
)
size_pattern = re.compile(r'^\s*\d+(\.\d+)?\s*(oz|ml|mL|fl\. oz|g|L|l|grams?|mL|ml|m\.L|m\.l)[^a-zA-Z0-9]*.*$', re.IGNORECASE)

# Apply a function to remove such values
df['type_raw'] = df['type_raw'].apply(
    lambda x: None if isinstance(x, str) and size_pattern.match(x.strip()) else x
)
df['type'] = None
non_null_type = df['type_raw'].notnull()
df.loc[non_null_type, 'type'] = encoder.fit_transform(df.loc[non_null_type, 'type_raw'])
df.drop(columns=['type_raw'], inplace=True)
```

```python
known_values = {
    'Dermask Intra Jet Wrinkless Solution', 'Watermelon + Vitamin E', 'Blemish', 'Hemp Mousse',
    'Grapefruit Peeling', 'Capture Youth Intense Rescue Age-Delay Revitalizing Oil-Serum',
    'Original', 'Dermask Intra Jet Firming Solution', 'Dark Spot',
    'Glow Booster Age-Delay Illuminating Serum', 'Dermask Ultra Jet Peeling Solution',
    'Combination + Oily Skin', 'Plump Filler Age-Delay Plumping Serum', 'Cucumber',
    'Mini NuFace / Sunset Blush', 'Rose', 'Blush', 'Peppermint', 'Vanilla Beige',
    'Cool Dewy', 'Vanilla', 'Berry', 'Minted Rose Lip Balm', 'Rosebud Salve',
    'Rosebud Salve in a Tube', 'Burgundy', 'Transforming Neutral Beige',
    'Translucent', 'Fuchsia', 'Jade', 'Clear', 'Coconut', 'Green', 'Pink', 'Watermelon',
    'Translucent Tone 1', 'Pearl Pink', 'Light', 'Light/Medium', 'Fair - Light', 'White',
    'Petal Pink', 'Caramel', 'Honey', 'Stone', 'White/Off-white', 'Navy', 'Fete',
    '772 Classic', 'Bliss', 'Peach Of Cake!', 'Lavender', 'New York', 'Unscented', 
    'Original', 'Papaya', 'Cucumber'
}
```

```python
def extract_quantity(value):
    if pd.isnull(value):
        return None
    if value in known_values:
        return None
    if 'oz' in value.lower():
        return None
    if 'ml' in value.lower():
        return None
    if '.' in value:
        return None
    return value

df['quantity_included_raw'] = df['variation_value'].apply(extract_quantity)
```

```python
def extract_first_number(value):
    if pd.isnull(value):
        return 1
    if value == '30 Treatments + 5 Bonus':
        return 35
    match = re.search(r'\d+', value)
    return int(match.group()) if match else 1

df['quantity_number'] = df['quantity_included_raw'].apply(extract_first_number)
df.drop(columns=['quantity_included_raw'], inplace=True)
```

### Reviewed text 

```python
def clean_text(text):
    if pd.isnull(text):
        return ""
    text = text.lower()                          # Lowercase
    text = re.sub(r'[^a-z0-9\s]', '', text)      # Remove punctuation and special characters
    text = re.sub(r'\s+', ' ', text).strip()     # Remove extra spaces
    return text

df['review_clean'] = df['review_text'].apply(clean_text)
del df['review_text']
```

```python
vectorizer = TfidfVectorizer(max_features=1000, ngram_range=(1,2))
X_review = vectorizer.fit_transform(df['review_clean'])
```
